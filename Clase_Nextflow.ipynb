```python
nextflow.enable.dsl=2

// Define all paths
params.input     = '/content/drive/MyDrive/Colab_Notebooks/gen_evol_2025/input/wc_nextflow.fasta'
params.outdir    = '/content/drive/MyDrive/Colab_Notebooks/gen_evol_2025/output'
params.scriptdir = '/content/drive/MyDrive/Colab_Notebooks/gen_evol_2025/scripts'

process countFasta_pros {
    publishDir params.outdir, mode: 'copy', overwrite: true

    input:
    path fasta_file

    output:
    path "${fasta_file.simpleName}.count.txt", emit: result

    script:
    """
    bash "${params.scriptdir}/countlines.sh" "${fasta_file}" > "${fasta_file.simpleName}.count.txt"
    """
}

workflow {
    results = countFasta_pros(params.input)
    results.view()
}
```
